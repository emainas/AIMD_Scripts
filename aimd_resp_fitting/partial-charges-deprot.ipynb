```python
def extract_esp_charges(filename, length):
    length = length
    all_charges = []
    with open(filename, 'r') as file:
        lines = file.readlines()

    i = 0
    while i < len(lines):
        if "ESP restrained charges:" in lines[i]:
            i += 3
            charges = []
            for j in range(length):
                if i + j < len(lines):
                    parts = lines[i + j].split()
                    if len(parts) >= 5:
                        charges.append(float(parts[4]))
            if len(charges) == length:
                all_charges.append(charges)
            i += length
        else:
            i += 1

    return np.array(all_charges)
```

```python
file_path1 = [
    'deprot/1/resp.out',
    'deprot/2/resp.out',
    'deprot/3/resp.out',
    'deprot/4/resp.out',
    'deprot/5/resp.out'
]

file_path2 = [
    'h76/1/resp.out',
    'h76/2/resp.out',
    'h76/3/resp.out',
    'h76/4/resp.out',
    'h76/5/resp.out',
    'h76/6/resp.out',
    'h76/7/resp.out'
]

file_path3 = [
    'h80/1/resp.out',
    'h80/2/resp.out',
    'h80/3/resp.out',
    'h80/4/resp.out',
    'h80/5/resp.out',
    'h80/6/resp.out'
]

file_path4 = [
    'twoprot/1/resp.out',
    'twoprot/2/resp.out',
    'twoprot/3/resp.out',
    'twoprot/4/resp.out',
    'twoprot/5/resp.out'
]

all_charges1 = []
all_charges2 = []
all_charges3 = []
all_charges4 = []

for path1 in file_path1:
    charges1 = extract_esp_charges(path1, 75)
    if charges1.size > 0:
        all_charges1.append(charges1)

for path2 in file_path2:
    charges2 = extract_esp_charges(path2, 76)
    if charges2.size > 0:
        all_charges2.append(charges2)

for path3 in file_path3:
    charges3 = extract_esp_charges(path3, 76)
    if charges3.size > 0:
        all_charges3.append(charges3)

for path4 in file_path4:
    charges4 = extract_esp_charges(path4, 77)
    if charges4.size > 0:
        all_charges4.append(charges4)

if all_charges1:
    combined_charges1 = np.concatenate(all_charges1, axis=0)
if all_charges2:
    combined_charges2 = np.concatenate(all_charges2, axis=0)
if all_charges3:
    combined_charges3 = np.concatenate(all_charges3, axis=0)
if all_charges4:
    combined_charges4 = np.concatenate(all_charges4, axis=0)

```

```python
import csv

means_stds1 = []
means_stds2 = []
means_stds3 = []
means_stds4 = []
for i in range(len(names)):
    data1 = combined_charges1[len(combined_charges1) // 2:, i]
    data2 = combined_charges2[len(combined_charges2) // 2:, i]
    data3 = combined_charges3[len(combined_charges3) // 2:, i]
    data4 = combined_charges4[len(combined_charges4) // 2:, i]
    #data = combined_charges1[:, i]
    mean1 = np.mean(data1)
    std1 = np.std(data1)
    mean2 = np.mean(data2)
    std2 = np.std(data2)
    mean3 = np.mean(data3)
    std3 = np.std(data3)
    mean4 = np.mean(data4)
    std4 = np.std(data4)
    means_stds1.append(f"{mean1:.4f} +/- {std1:.4f}")
    means_stds2.append(f"{mean2:.4f} +/- {std2:.4f}")
    means_stds3.append(f"{mean3:.4f} +/- {std3:.4f}")
    means_stds4.append(f"{mean4:.4f} +/- {std4:.4f}")

table = []
for name, mean_std1, deprot, mean_std2, h76, mean_std3, h80 , mean_std4, twoprot in zip(names, means_stds1, deprot_old, means_stds2, h76_old, means_stds3, h80_old, means_stds4, twoprot_old):
    table.append(f"{name: <6} | {mean_std1: <20} | {deprot: .6f} | {mean_std2: <20} | {h76: .6f} | {mean_std3: <20} | {h80: .6f} | {mean_std4: <20} | {twoprot: .6f}")

print(r'Mask   | Deprot new: μ +/- σ  | Deprot-old| H76 new: μ +/- σ     |  H76 old  | H80 new: μ +/- σ     |  H80 old | Twoprot new: μ +/- σ     |  Twoprot old')
print("-----------------------------------------------------------------------------------------------------------------------------------------------------------")
for row in table:
    print(row)
```

```python
with open('data_table.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Mask', 'Deprot new: μ +/- σ', 'Deprot old', 'H76 new: μ +/- σ', 'H76 old', 'H80 new: μ +/- σ', 'H80 old', 'Twoprot new: μ +/- σ', 'Twoprot old'])

    for name, mean_std1, deprot, mean_std2, h76, mean_std3, h80, mean_std4, twoprot in zip(names, means_stds1, deprot_old, means_stds2, h76_old, means_stds3, h80_old, means_stds4, twoprot_old):
        writer.writerow([name, mean_std1, f"{deprot:.6f}", mean_std2, f"{h76:.6f}", mean_std3, f"{h80:.6f}", mean_std4, f"{twoprot:.6f}"])

print("CSV file 'data_table.csv' has been created successfully.")
```
